```python
# Now seperating the input column & out columns for further calculation

df1= df_new.iloc[:,0:-1].values
df2= df_new.iloc[:, -1].values

df1=pd.DataFrame(df1)
df2=pd.DataFrame(df2)

#Assigning the x & y cordinate value for train, test, split function
x=df1
y=df2

print(x)
print(x.shape)
print()
print(y)
print(y.shape)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn import linear_model
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.22, random_state=42)
```

```python
def maxr2_score(regr, x, y):
    max_r_score=0
    for r_state in range(42, 200):
        x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = r_state, test_size=0.20)
        regr.fit(x_train, y_train)
        y_pred = regr.predict(x_test)
        r2_scr=r2_score(y_test, y_pred)
        ("r2 score corresponding to ", r_state, " is ", r2_scr)
        if r2_scr>max_r_score:
            max_r_score=r2_scr
            final_r_state=r_state
    print("max r2 score corresponding to ", final_r_state, " is ", max_r_score)
    return final_r_state
```

# Linear Regression

```python
#Lets use linear regression and check max r2 score corresponding to different random states

from sklearn.linear_model import LinearRegression

lreg=LinearRegression()
r_state=maxr2_score(lreg, x, y)
```

```python
from sklearn.model_selection import cross_val_score
print("Mean r2 score for Linear Regression: ", cross_val_score(lreg, x, y, cv=5, scoring="r2").mean())
print()
print("Standard deviation in r2 score for Linear Regression: ", cross_val_score(lreg, x, y, cv=5, scoring="r2").std())
```

# KNeighbors Regressor

```python
#Lets use grid_search to find optimal value of n_neighbors for KNN model

from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsRegressor

neighbors={"n_neighbors":range(1,30)}
knr=KNeighborsRegressor()
gknr = GridSearchCV(knr, neighbors, cv=10)
gknr.fit(x, y)
gknr.best_params_
```

```python
#Lets use KNN regression and check max r2 score corresponding to different random states

knr=KNeighborsRegressor(n_neighbors=29)
r_state=maxr2_score(knr, x, y)
```

```python
#Lets check the mean r2 score KNeighbors Regressor

print("mean r2 score for KNN Regression: ", cross_val_score(knr, x, y, cv=5, scoring="r2").mean())
print()
print("standard deviation in r2 score for KNN Regression: ", cross_val_score(knr, x, y, cv=5, scoring="r2").std())
```

# Random Forest Regressor

```python
#Lets use Random Forest Regressor regression and check max r2 score 

from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor(n_estimators=30)
r_state=maxr2_score(rfr, x, y)
```

```python
#Lets check the mean r2 score for Random Forest Regressor

print("Mean r2 score for RandomForest Regression: ", cross_val_score(rfr, x, y, cv=5, scoring="r2").mean())
print()
print("Standard deviation in r2 score for Linear Regression: ", cross_val_score(rfr, x, y, cv=5, scoring="r2").std())
print()
```

# Lasso regression

```python
#Lets check Lasso regression and find best value of alpha
from sklearn.linear_model import Lasso
lsreg=Lasso()
parameters={"alpha" : [0.001, 0.01, 0.1, 1]}
clf = GridSearchCV(lsreg, parameters, cv=10)
clf.fit(x, y)
clf.best_params_
```

```python
#Lets check max r2 score when we use Lasso
lsreg=Lasso(alpha=1)
r_state=maxr2_score(lsreg, x, y)
```

```python
#Lets use cross val score with Lasso

print("Mean r2 score for Lasso regression: ", cross_val_score(lsreg, x, y, cv=5, scoring="r2").mean())
print()
print("standard deviation in r2 score for Lasso Regression: ", cross_val_score(lsreg, x, y, cv=5, scoring="r2").std())
```

# Finalysing the Model

```python
#we tried several the models and till now RandomForest Regression is the best
#random state corresponding to highest r2-score is 111

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 111, test_size=0.20)
rfr=RandomForestRegressor()
rfr.fit(x_train, y_train)
y_pred=rfr.predict(x_test)
```

```python
#Lets find the rmse and r2_score using sklearn.metrics
import numpy as np
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
print("RMSE is: ", np.sqrt(mean_squared_error(y_test, y_pred)))
print("re_score is ", r2_score(y_test, y_pred))
```
